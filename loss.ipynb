```python
import numpy as np
import matplotlib.pyplot as plt
import joblib
```

```python
loss = joblib.load('loss')
loss[0]
```

```python
new_loss = [loss[l].detach().cpu() for l in range(len(loss)) if l%400==0]
plt.plot(new_loss)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
```
